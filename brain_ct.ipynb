```python
# این سلول رو اجرا کن تا آرگومان‌های جدید به train.py اضافه بشه
%%writefile /content/ct-segmentation-contest/train_enhanced.py
#!/usr/bin/env python3
"""
Brain CT Triage Challenge - Enhanced Training Script
با قابلیت‌های پیشرفته برای مسابقه
"""

import torch
import torch.nn as nn
import numpy as np
from pathlib import Path
import pandas as pd
import json
import argparse
import sys
import os
from sklearn.model_selection import train_test_split, KFold
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# اضافه کردن مسیر src به sys.path
sys.path.append(str(Path(__file__).parent / 'src'))

from src.datasets.ct_dataset import CTDataset3D
from src.models.advanced_unet import MultiTaskUNet, MultiTaskLoss
from src.preprocessing.dicom_processor import DICOMProcessor, AnnotationProcessor

# ========== اضافه کردن ماژول‌های جدید ==========
try:
    from src.models.attention_unet import AttentionUNet
    from src.models.deeplabv3 import DeepLabV3Plus
    print("✅ ماژول‌های پیشرفته پیدا شدن")
except ImportError:
    print("⚠️ ماژول‌های پیشرفته موجود نیستن، از مدل پایه استفاده می‌شه")

try:
    import albumentations as A
    from albumentations.pytorch import ToTensorV2
    AUG_AVAILABLE = True
except ImportError:
    AUG_AVAILABLE = False
    print("⚠️ albumentations نصب نیست. اجرا کن: !pip install albumentations")
# ==============================================


class BrainCTTriageTrainer:
    """Trainer class for Brain CT Triage Challenge"""

    def __init__(self, model, train_loader, val_loader, config):
        self.model = model
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.config = config
        self.device = torch.device(config['device'])

        # Move model to device
        self.model = self.model.to(self.device)

        # ========== Loss ترکیبی پیشرفته ==========
        self.use_advanced_loss = config.get('use_advanced_loss', False)
        if self.use_advanced_loss:
            try:
                from src.losses.combined_loss import CombinedLoss
                self.criterion = CombinedLoss(
                    seg_weight=config.get('seg_weight', 1.0),
                    mls_weight=config.get('mls_weight', 0.5),
                    fracture_weight=config.get('fracture_weight', 0.3),
                    dice_weight=config.get('dice_weight', 0.5),
                    focal_weight=config.get('focal_weight', 0.5),
                    boundary_weight=config.get('boundary_weight', 0.2)
                )
                print("✅ استفاده از Loss ترکیبی پیشرفته")
            except ImportError:
                print("⚠️ CombinedLoss موجود نیست، استفاده از loss پایه")
                self.criterion = MultiTaskLoss(
                    seg_weight=config.get('seg_weight', 1.0),
                    mls_weight=config.get('mls_weight', 0.5),
                    fracture_weight=config.get('fracture_weight', 0.3)
                )
        else:
            self.criterion = MultiTaskLoss(
                seg_weight=config.get('seg_weight', 1.0),
                mls_weight=config.get('mls_weight', 0.5),
                fracture_weight=config.get('fracture_weight', 0.3)
            )
        # ==========================================

        # Optimizer
        self.optimizer = torch.optim.AdamW(
            self.model.parameters(),
            lr=config['learning_rate'],
            weight_decay=config.get('weight_decay', 1e-5)
        )

        # ========== Scheduler پیشرفته ==========
        scheduler_type = config.get('scheduler', 'cosine')
        if scheduler_type == 'cosine':
            self.scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
                self.optimizer,
                T_max=config['num_epochs'],
                eta_min=1e-6
            )
        elif scheduler_type == 'cosine_warm':
            try:
                self.scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(
                    self.optimizer,
                    T_0=10,
                    T_mult=2,
                    eta_min=1e-6
                )
                print("✅ استفاده از CosineAnnealingWarmRestarts")
            except:
                self.scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
                    self.optimizer, T_max=config['num_epochs']
                )
        elif scheduler_type == 'reduce':
            self.scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
                self.optimizer, mode='min', factor=0.5, patience=5
            )
        else:
            self.scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
                self.optimizer, T_max=config['num_epochs']
            )
        # ==========================================

        # Early stopping
        self.patience = config.get('patience', 15)
        self.best_val_loss = float('inf')
        self.patience_counter = 0
        self.best_model_state = None

        # Metrics tracking
        self.train_losses = []
        self.val_losses = []

        # Checkpoint directory
        self.checkpoint_dir = Path(config.get('checkpoint_dir', 'checkpoints'))
        self.checkpoint_dir.mkdir(exist_ok=True)

    def train_epoch(self):
        """Train for one epoch"""
        self.model.train()
        epoch_loss = 0

        pbar = tqdm(self.train_loader, desc="Training", leave=False)
        for batch_idx, batch in enumerate(pbar):
            # Move data to device
            images = batch['image'].to(self.device)
            seg_targets = batch['segmentation'].to(self.device)
            mls_targets = batch['mls_mm'].to(self.device)
            fracture_targets = batch['fracture_prob'].to(self.device)

            # Forward pass
            outputs = self.model(images)

            # Prepare targets
            targets = {
                'segmentation': seg_targets,
                'mls': mls_targets,
                'fracture': fracture_targets
            }

            # Calculate loss
            losses = self.criterion(outputs, targets)
            loss = losses['total_loss'] if isinstance(losses, dict) else losses

            # Backward pass
            self.optimizer.zero_grad()
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)

            self.optimizer.step()

            # Update metrics
            epoch_loss += loss.item()

            # Update progress bar
            if isinstance(losses, dict):
                pbar.set_postfix({
                    'loss': f"{loss.item():.4f}",
                    'seg': f"{losses['seg_loss'].item():.4f}" if 'seg_loss' in losses else 'N/A'
                })
            else:
                pbar.set_postfix({'loss': f"{loss.item():.4f}"})

        # Average losses
        n_batches = len(self.train_loader)
        return epoch_loss / n_batches

    def validate(self):
        """Validate the model"""
        self.model.eval()
        epoch_loss = 0

        with torch.no_grad():
            pbar = tqdm(self.val_loader, desc="Validation", leave=False)
            for batch in pbar:
                # Move data to device
                images = batch['image'].to(self.device)
                seg_targets = batch['segmentation'].to(self.device)
                mls_targets = batch['mls_mm'].to(self.device)
                fracture_targets = batch['fracture_prob'].to(self.device)

                # Forward pass
                outputs = self.model(images)

                # Prepare targets
                targets = {
                    'segmentation': seg_targets,
                    'mls': mls_targets,
                    'fracture': fracture_targets
                }

                # Calculate loss
                losses = self.criterion(outputs, targets)
                loss = losses['total_loss'] if isinstance(losses, dict) else losses

                # Update metrics
                epoch_loss += loss.item()

        # Average losses
        n_batches = len(self.val_loader)
        return epoch_loss / n_batches

    def train(self, num_epochs):
        """Main training loop"""
        print("\n" + "="*60)
        print("STARTING TRAINING")
        print("="*60)

        for epoch in range(1, num_epochs + 1):
            print(f"\nEpoch {epoch}/{num_epochs}")
            print("-"*40)

            # Train
            train_loss = self.train_epoch()

            # Validate
            val_loss = self.validate()

            # Update scheduler
            if isinstance(self.scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                self.scheduler.step(val_loss)
            else:
                self.scheduler.step()

            # Log metrics
            self.train_losses.append(train_loss)
            self.val_losses.append(val_loss)

            # Print metrics
            print(f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
            lr = self.optimizer.param_groups[0]['lr']
            print(f"LR: {lr:.2e}")

            # Save checkpoint
            if epoch % 5 == 0:
                self.save_checkpoint(epoch, val_loss)

            # Early stopping
            if val_loss < self.best_val_loss:
                self.best_val_loss = val_loss
                self.patience_counter = 0
                self.best_model_state = self.model.state_dict().copy()

                # Save best model
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': self.model.state_dict(),
                    'optimizer_state_dict': self.optimizer.state_dict(),
                    'val_loss': val_loss,
                    'config': self.config
                }, self.checkpoint_dir / 'best_model.pth')
                print("✓ New best model saved!")
            else:
                self.patience_counter += 1

            if self.patience_counter >= self.patience:
                print(f"\nEarly stopping triggered after {epoch} epochs")
                break

        # Load best model
        if self.best_model_state is not None:
            self.model.load_state_dict(self.best_model_state)

        return self.train_losses, self.val_losses

    def save_checkpoint(self, epoch, val_loss):
        """Save checkpoint"""
        checkpoint_path = self.checkpoint_dir / f'checkpoint_epoch_{epoch}.pth'
        torch.save({
            'epoch': epoch,
            'model_state_dict': self.model.state_dict(),
            'optimizer_state_dict': self.optimizer.state_dict(),
            'val_loss': val_loss,
            'config': self.config
        }, checkpoint_path)


def load_metadata(data_dir: Path) -> pd.DataFrame:
    """Load metadata with proper column detection"""
    metadata_path = data_dir / "raw" / "training_df.pkl"

    try:
        if metadata_path.exists():
            metadata = pd.read_pickle(metadata_path)
            print(f"✓ Loaded metadata with shape: {metadata.shape}")

            # Try to find the series ID column
            possible_id_columns = ['dicom_series.id', 'series_id', 'id', 'StudyInstanceUID', 'PatientID']
            id_column = None

            for col in possible_id_columns:
                if col in metadata.columns:
                    id_column = col
                    break

            if id_column:
                metadata.set_index(id_column, inplace=True)
                print(f"✓ Set index to '{id_column}'")
            else:
                print(f"⚠️ No ID column found. Using first column as index.")
                if metadata.shape[1] > 0:
                    metadata.set_index(metadata.columns[0], inplace=True)

            return metadata
        else:
            print(f"✗ Metadata not found at {metadata_path}")
            return None
    except Exception as e:
        print(f"✗ Error loading metadata: {e}")
        return None


def prepare_data_paths(data_dir: Path, metadata: pd.DataFrame, num_samples: int = None) -> list:
    """Prepare data paths with proper ID matching"""
    training_dir = data_dir / "raw" / "training"

    if not training_dir.exists():
        print(f"✗ Training directory not found: {training_dir}")
        return []

    # Get metadata IDs as strings
    metadata_ids = set(metadata.index.astype(str))
    print(f"   Metadata has {len(metadata_ids)} unique IDs")

    # List DICOM folders
    dicom_folders = []
    missing_ids = []

    all_folders = list(training_dir.iterdir())
    if num_samples:
        all_folders = all_folders[:num_samples]

    for folder in tqdm(all_folders, desc="Scanning DICOM folders"):
        if folder.is_dir():
            folder_name = folder.name
            dicom_files = list(folder.glob("*.dcm"))

            if dicom_files:
                if folder_name in metadata_ids:
                    dicom_folders.append(folder)
                else:
                    missing_ids.append(folder_name)

    print(f"✓ Found {len(dicom_folders)} valid DICOM folders")
    return dicom_folders


def prepare_processed_data(data_dir: Path, dicom_folders: list, force_reprocess: bool = False, verbose: bool = False) -> list:
    """Process and save data"""
    processed_dir = data_dir / "processed"
    processed_dir.mkdir(exist_ok=True)

    processed_paths = []
    skipped_count = 0

    for dicom_folder in tqdm(dicom_folders, desc="Processing data"):
        series_id = dicom_folder.name

        # Save paths
        volume_path = processed_dir / f"{series_id}_volume.npy"
        seg_path = processed_dir / f"{series_id}_segmentation.npy"

        if volume_path.exists() and seg_path.exists() and not force_reprocess:
            # Already processed
            processed_paths.append({
                'series_id': series_id,
                'volume_path': str(volume_path),
                'seg_path': str(seg_path),
                'dicom_folder': str(dicom_folder)
            })
            skipped_count += 1
            continue

        # This is just for tracking - 실제 processing은 CTDataset3D에서 함
        processed_paths.append({
            'series_id': series_id,
            'volume_path': str(volume_path),
            'seg_path': str(seg_path),
            'dicom_folder': str(dicom_folder)
        })

    print(f"\n✓ Data processing complete:")
    print(f"   - Skipped (already processed): {skipped_count}")
    print(f"   - Total available: {len(processed_paths)} series")

    return processed_paths


def get_augmentation_pipeline():
    """Get data augmentation pipeline"""
    if not AUG_AVAILABLE:
        return None

    return A.Compose([
        A.RandomRotate90(p=0.5),
        A.Flip(p=0.5),
        A.Transpose(p=0.5),
        A.GaussNoise(var_limit=(0.01, 0.05), p=0.2),
        A.RandomBrightnessContrast(brightness_limit=0.1, contrast_limit=0.1, p=0.3),
        A.RandomGamma(gamma_limit=(80, 120), p=0.2),
        A.ElasticTransform(alpha=1, sigma=50, alpha_affine=50, p=0.1),
    ])


def main():
    """Main training function"""
    parser = argparse.ArgumentParser(description='Brain CT Triage Challenge Enhanced Training')
    parser.add_argument('--mode', type=str, default='normal',
                       choices=['normal', 'cv'],
                       help='Training mode: normal or cross-validation')
    parser.add_argument('--epochs', type=int, default=50,
                       help='Number of epochs')
    parser.add_argument('--batch_size', type=int, default=2,
                       help='Batch size')
    parser.add_argument('--lr', type=float, default=1e-4,
                       help='Learning rate')
    parser.add_argument('--num_samples', type=int, default=None,
                       help='Number of samples to use (for testing)')
    parser.add_argument('--force_reprocess', action='store_true',
                       help='Force reprocessing of all data')
    parser.add_argument('--verbose', action='store_true',
                       help='Print verbose output')

    # ========== آرگومان‌های جدید ==========
    parser.add_argument('--model_type', type=str, default='unet',
                       choices=['unet', 'attention_unet', 'deeplabv3'],
                       help='Model architecture to use')
    parser.add_argument('--use_attention', action='store_true',
                       help='Use attention mechanism')
    parser.add_argument('--use_augmentation', action='store_true',
                       help='Use data augmentation')
    parser.add_argument('--scheduler', type=str, default='cosine',
                       choices=['cosine', 'cosine_warm', 'reduce'],
                       help='Learning rate scheduler')
    parser.add_argument('--use_advanced_loss', action='store_true',
                       help='Use combined loss (Dice + Focal + Boundary)')
    parser.add_argument('--folds', type=int, default=5,
                       help='Number of folds for cross-validation')
    # ======================================

    args = parser.parse_args()

    print("\n" + "="*60)
    print("IAAA COMPETITION 2026 - BRAIN CT TRIAGE CHALLENGE")
    print("="*60)
    print(f"Mode: {args.mode}")
    print(f"Epochs: {args.epochs}")
    print(f"Batch size: {args.batch_size}")
    print(f"Learning rate: {args.lr}")
    print(f"Model type: {args.model_type}")
    print(f"Use attention: {args.use_attention}")
    print(f"Use augmentation: {args.use_augmentation}")
    print(f"Scheduler: {args.scheduler}")
    print(f"Use advanced loss: {args.use_advanced_loss}")

    if args.num_samples:
        print(f"Using {args.num_samples} samples (debug mode)")

    # Check for augmentations
    if args.use_augmentation and not AUG_AVAILABLE:
        print("⚠️ albumentations not installed. Installing...")
        !pip install albumentations
        print("✅ albumentations installed. Please restart the runtime and run again.")
        return

    # Configuration
    config = {
        # Model architecture
        'model_type': args.model_type,
        'use_attention': args.use_attention,
        'model_params': {
            'in_channels': 1,
            'seg_classes': 6,
            'init_features': 32
        },

        # Training hyperparameters
        'learning_rate': args.lr,
        'weight_decay': 1e-5,
        'batch_size': args.batch_size,
        'num_epochs': args.epochs,
        'patience': 15,

        # Loss weights
        'seg_weight': 1.0,
        'mls_weight': 0.5,
        'fracture_weight': 0.3,

        # Advanced loss weights
        'dice_weight': 0.5,
        'focal_weight': 0.5,
        'boundary_weight': 0.2,

        # Dataset parameters
        'dataset_params': {
            'target_size': (256, 256),
            'num_slices': 32,
            'augmentation': args.use_augmentation
        },

        # System
        'device': 'cuda' if torch.cuda.is_available() else 'cpu',
        'num_workers': 2,
        'pin_memory': False,
        'use_wandb': False,
        'checkpoint_dir': 'checkpoints',

        # New params
        'scheduler': args.scheduler,
        'use_advanced_loss': args.use_advanced_loss
    }

    print("\n" + "="*60)
    print("BRAIN CT TRIAGE CHALLENGE - TRAINING")
    print("="*60)

    # Print configuration
    print("\n📋 CONFIGURATION:")
    for key, value in config.items():
        if isinstance(value, dict):
            print(f"   {key}:")
            for k, v in value.items():
                print(f"      {k}: {v}")
        else:
            print(f"   {key}: {value}")

    # 1. Load metadata
    print("\n📊 LOADING METADATA...")
    data_dir = Path("data")
    metadata = load_metadata(data_dir)

    if metadata is None:
        print("✗ No metadata found. Exiting...")
        return

    # 2. Prepare data paths
    print("\n🔍 PREPARING DATA PATHS...")
    dicom_folders = prepare_data_paths(data_dir, metadata, args.num_samples)

    if not dicom_folders:
        print("✗ No valid DICOM folders found. Exiting...")
        return

    # 3. Process data
    print("\n⚙️ PROCESSING DATA...")
    processed_paths = prepare_processed_data(data_dir, dicom_folders, args.force_reprocess, args.verbose)

    if len(processed_paths) == 0:
        print("✗ No data processed successfully. Exiting...")
        return

    if args.mode == 'normal':
        # 4. Train/val split
        print("\n📦 SPLITTING DATA...")
        train_paths, val_paths = train_test_split(
            processed_paths,
            test_size=0.15,
            random_state=42,
            shuffle=True
        )

        print(f"   Train: {len(train_paths)} series")
        print(f"   Val: {len(val_paths)} series")

        # 5. Create datasets with augmentation if requested
        print("\n🎯 CREATING DATASETS...")

        # Get augmentation pipeline
        aug_transform = get_augmentation_pipeline() if args.use_augmentation else None

        train_dataset = CTDataset3D(
            dicom_dirs=[p['dicom_folder'] for p in train_paths],
            annotation_dirs=[p['dicom_folder'] for p in train_paths],
            metadata_df=metadata,
            is_training=True,
            processed_dir="data/processed",
            slice_based=True,
            transform=aug_transform,
            **config['dataset_params']
        )

        val_dataset = CTDataset3D(
            dicom_dirs=[p['dicom_folder'] for p in val_paths],
            annotation_dirs=[p['dicom_folder'] for p in val_paths],
            metadata_df=metadata,
            is_training=False,
            processed_dir="data/processed",
            slice_based=True,
            **config['dataset_params']
        )

        # 6. Create dataloaders
        train_loader = torch.utils.data.DataLoader(
            train_dataset,
            batch_size=config['batch_size'],
            shuffle=True,
            num_workers=config['num_workers'],
            pin_memory=config['pin_memory'],
            drop_last=True
        )

        val_loader = torch.utils.data.DataLoader(
            val_dataset,
            batch_size=config['batch_size'],
            shuffle=False,
            num_workers=config['num_workers'],
            pin_memory=config['pin_memory']
        )

        # 7. Create model
        print("\n🧠 CREATING MODEL...")

        if args.model_type == 'attention_unet' and args.use_attention:
            try:
                from src.models.attention_unet import AttentionUNet
                model = AttentionUNet(**config['model_params'])
                print("✅ Using Attention U-Net")
            except ImportError:
                print("⚠️ Attention U-Net not available, using MultiTaskUNet")
                model = MultiTaskUNet(**config['model_params'])
        elif args.model_type == 'deeplabv3':
            try:
                from src.models.deeplabv3 import DeepLabV3Plus
                model = DeepLabV3Plus(**config['model_params'])
                print("✅ Using DeepLabV3+")
            except ImportError:
                print("⚠️ DeepLabV3+ not available, using MultiTaskUNet")
                model = MultiTaskUNet(**config['model_params'])
        else:
            model = MultiTaskUNet(**config['model_params'])
            print("✅ Using MultiTaskUNet")

        total_params = sum(p.numel() for p in model.parameters())
        trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
        print(f"   Total parameters: {total_params:,}")
        print(f"   Trainable parameters: {trainable_params:,}")

        # 8. Create trainer and train
        print("\n🚀 STARTING TRAINING...")
        trainer = BrainCTTriageTrainer(
            model=model,
            train_loader=train_loader,
            val_loader=val_loader,
            config=config
        )

        train_losses, val_losses = trainer.train(config['num_epochs'])

        # 9. Save final model
        print("\n💾 SAVING FINAL MODEL...")
        weights_dir = Path("weights")
        weights_dir.mkdir(exist_ok=True)

        model_name = f"{args.model_type}"
        if args.use_attention:
            model_name += "_attention"
        if args.use_augmentation:
            model_name += "_aug"
        if args.use_advanced_loss:
            model_name += "_advloss"

        final_model_path = weights_dir / f"{model_name}.pth"
        torch.save({
            'model_state_dict': trainer.best_model_state,
            'config': config,
            'train_losses': train_losses,
            'val_losses': val_losses,
            'best_val_loss': trainer.best_val_loss,
            'epoch': len(train_losses)
        }, final_model_path)

        print(f"✓ Model saved to {final_model_path}")
        print(f"✓ Best validation loss: {trainer.best_val_loss:.4f}")

        # 10. Save training config
        config_path = weights_dir / f"{model_name}_config.json"
        with open(config_path, 'w') as f:
            serializable_config = {}
            for k, v in config.items():
                if isinstance(v, Path):
                    serializable_config[k] = str(v)
                elif isinstance(v, tuple):
                    serializable_config[k] = list(v)
                elif isinstance(v, dict):
                    serializable_config[k] = {kk: str(vv) if isinstance(vv, Path) else vv
                                             for kk, vv in v.items()}
                else:
                    serializable_config[k] = v
            json.dump(serializable_config, f, indent=2)

        print(f"✓ Config saved to {config_path}")

        print("\n" + "="*60)
        print("✅ TRAINING COMPLETED SUCCESSFULLY!")
        print("="*60)

    else:  # Cross-validation mode
        print("\n📊 CROSS-VALIDATION MODE")
        print("   (To be implemented in next version)")


if __name__ == "__main__":
    main()
```

```python
%%writefile /content/ct-segmentation-contest/train_enhanced.py
#!/usr/bin/env python3
"""
Brain CT Triage Challenge - Enhanced Training Script
با قابلیت‌های پیشرفته برای مسابقه
"""

import torch
import torch.nn as nn
import numpy as np
from pathlib import Path
import pandas as pd
import json
import argparse
import sys
import os
from sklearn.model_selection import train_test_split, KFold
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# اضافه کردن مسیر src به sys.path
sys.path.append(str(Path(__file__).parent / 'src'))

from src.datasets.ct_dataset import CTDataset3D
from src.models.advanced_unet import MultiTaskUNet, MultiTaskLoss
from src.preprocessing.dicom_processor import DICOMProcessor, AnnotationProcessor

# ========== اضافه کردن ماژول‌های جدید ==========
try:
    from src.models.attention_unet import AttentionUNet
    print("✅ ماژول Attention UNet پیدا شد")
except ImportError:
    print("⚠️ ماژول Attention UNet موجود نیست، از مدل پایه استفاده می‌شه")

try:
    from src.models.deeplabv3 import DeepLabV3Plus
    print("✅ ماژول DeepLabV3+ پیدا شد")
except ImportError:
    print("⚠️ ماژول DeepLabV3+ موجود نیست، از مدل پایه استفاده می‌شه")

try:
    import albumentations as A
    AUG_AVAILABLE = True
    print("✅ albumentations نصب است")
except ImportError:
    AUG_AVAILABLE = False
    print("⚠️ albumentations نصب نیست. اجرا کن: !pip install albumentations")
# ==============================================


class BrainCTTriageTrainer:
    """Trainer class for Brain CT Triage Challenge"""

    def __init__(self, model, train_loader, val_loader, config):
        self.model = model
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.config = config
        self.device = torch.device(config['device'])

        # Move model to device
        self.model = self.model.to(self.device)

        # Loss function
        self.criterion = MultiTaskLoss(
            seg_weight=config.get('seg_weight', 1.0),
            mls_weight=config.get('mls_weight', 0.5),
            fracture_weight=config.get('fracture_weight', 0.3)
        )

        # Optimizer
        self.optimizer = torch.optim.AdamW(
            self.model.parameters(),
            lr=config['learning_rate'],
            weight_decay=config.get('weight_decay', 1e-5)
        )

        # Scheduler
        scheduler_type = config.get('scheduler', 'cosine')
        if scheduler_type == 'cosine_warm':
            self.scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(
                self.optimizer,
                T_0=10,
                T_mult=2,
                eta_min=1e-6
            )
            print("✅ Using CosineAnnealingWarmRestarts")
        else:
            self.scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
                self.optimizer,
                T_max=config['num_epochs'],
                eta_min=1e-6
            )

        # Early stopping
        self.patience = config.get('patience', 15)
        self.best_val_loss = float('inf')
        self.patience_counter = 0
        self.best_model_state = None

        # Metrics tracking
        self.train_losses = []
        self.val_losses = []

        # Checkpoint directory
        self.checkpoint_dir = Path(config.get('checkpoint_dir', 'checkpoints'))
        self.checkpoint_dir.mkdir(exist_ok=True)

    def train_epoch(self):
        """Train for one epoch"""
        self.model.train()
        epoch_loss = 0

        pbar = tqdm(self.train_loader, desc="Training", leave=False)
        for batch_idx, batch in enumerate(pbar):
            # Move data to device
            images = batch['image'].to(self.device)
            seg_targets = batch['segmentation'].to(self.device)
            mls_targets = batch['mls_mm'].to(self.device)
            fracture_targets = batch['fracture_prob'].to(self.device)

            # Forward pass
            outputs = self.model(images)

            # Prepare targets
            targets = {
                'segmentation': seg_targets,
                'mls': mls_targets,
                'fracture': fracture_targets
            }

            # Calculate loss
            losses = self.criterion(outputs, targets)
            loss = losses['total_loss']

            # Backward pass
            self.optimizer.zero_grad()
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)

            self.optimizer.step()

            # Update metrics
            epoch_loss += loss.item()

            # Update progress bar
            pbar.set_postfix({'loss': f"{loss.item():.4f}"})

        # Average losses
        n_batches = len(self.train_loader)
        return epoch_loss / n_batches

    def validate(self):
        """Validate the model"""
        self.model.eval()
        epoch_loss = 0

        with torch.no_grad():
            pbar = tqdm(self.val_loader, desc="Validation", leave=False)
            for batch in pbar:
                # Move data to device
                images = batch['image'].to(self.device)
                seg_targets = batch['segmentation'].to(self.device)
                mls_targets = batch['mls_mm'].to(self.device)
                fracture_targets = batch['fracture_prob'].to(self.device)

                # Forward pass
                outputs = self.model(images)

                # Prepare targets
                targets = {
                    'segmentation': seg_targets,
                    'mls': mls_targets,
                    'fracture': fracture_targets
                }

                # Calculate loss
                losses = self.criterion(outputs, targets)
                loss = losses['total_loss']

                # Update metrics
                epoch_loss += loss.item()

        # Average losses
        n_batches = len(self.val_loader)
        return epoch_loss / n_batches

    def train(self, num_epochs):
        """Main training loop"""
        print("\n" + "="*60)
        print("STARTING TRAINING")
        print("="*60)

        for epoch in range(1, num_epochs + 1):
            print(f"\nEpoch {epoch}/{num_epochs}")
            print("-"*40)

            # Train
            train_loss = self.train_epoch()

            # Validate
            val_loss = self.validate()

            # Update scheduler
            if isinstance(self.scheduler, torch.optim.lr_scheduler.CosineAnnealingWarmRestarts):
                self.scheduler.step()
            else:
                self.scheduler.step()

            # Log metrics
            self.train_losses.append(train_loss)
            self.val_losses.append(val_loss)

            # Print metrics
            print(f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
            lr = self.optimizer.param_groups[0]['lr']
            print(f"LR: {lr:.2e}")

            # Save checkpoint
            if epoch % 5 == 0:
                self.save_checkpoint(epoch, val_loss)

            # Early stopping
            if val_loss < self.best_val_loss:
                self.best_val_loss = val_loss
                self.patience_counter = 0
                self.best_model_state = self.model.state_dict().copy()

                # Save best model
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': self.model.state_dict(),
                    'optimizer_state_dict': self.optimizer.state_dict(),
                    'val_loss': val_loss,
                    'config': self.config
                }, self.checkpoint_dir / 'best_model.pth')
                print("✓ New best model saved!")
            else:
                self.patience_counter += 1

            if self.patience_counter >= self.patience:
                print(f"\nEarly stopping triggered after {epoch} epochs")
                break

        # Load best model
        if self.best_model_state is not None:
            self.model.load_state_dict(self.best_model_state)

        return self.train_losses, self.val_losses

    def save_checkpoint(self, epoch, val_loss):
        """Save checkpoint"""
        checkpoint_path = self.checkpoint_dir / f'checkpoint_epoch_{epoch}.pth'
        torch.save({
            'epoch': epoch,
            'model_state_dict': self.model.state_dict(),
            'optimizer_state_dict': self.optimizer.state_dict(),
            'val_loss': val_loss,
            'config': self.config
        }, checkpoint_path)


def load_metadata(data_dir: Path) -> pd.DataFrame:
    """Load metadata with proper column detection"""
    metadata_path = data_dir / "raw" / "training_df.pkl"

    try:
        if metadata_path.exists():
            metadata = pd.read_pickle(metadata_path)
            print(f"✓ Loaded metadata with shape: {metadata.shape}")

            # Try to find the series ID column
            possible_id_columns = ['dicom_series.id', 'series_id', 'id', 'StudyInstanceUID', 'PatientID']
            id_column = None

            for col in possible_id_columns:
                if col in metadata.columns:
                    id_column = col
                    break

            if id_column:
                metadata.set_index(id_column, inplace=True)
                print(f"✓ Set index to '{id_column}'")
            else:
                print(f"⚠️ No ID column found. Using first column as index.")
                if metadata.shape[1] > 0:
                    metadata.set_index(metadata.columns[0], inplace=True)

            return metadata
        else:
            print(f"✗ Metadata not found at {metadata_path}")
            return None
    except Exception as e:
        print(f"✗ Error loading metadata: {e}")
        return None


def prepare_data_paths(data_dir: Path, metadata: pd.DataFrame, num_samples: int = None) -> list:
    """Prepare data paths with proper ID matching"""
    training_dir = data_dir / "raw" / "training"

    if not training_dir.exists():
        print(f"✗ Training directory not found: {training_dir}")
        return []

    # Get metadata IDs as strings
    metadata_ids = set(metadata.index.astype(str))
    print(f"   Metadata has {len(metadata_ids)} unique IDs")

    # List DICOM folders
    dicom_folders = []

    all_folders = list(training_dir.iterdir())
    if num_samples:
        all_folders = all_folders[:num_samples]

    for folder in tqdm(all_folders, desc="Scanning DICOM folders"):
        if folder.is_dir():
            folder_name = folder.name
            dicom_files = list(folder.glob("*.dcm"))

            if dicom_files:
                if folder_name in metadata_ids:
                    dicom_folders.append(folder)

    print(f"✓ Found {len(dicom_folders)} valid DICOM folders")
    return dicom_folders


def prepare_processed_data(data_dir: Path, dicom_folders: list, force_reprocess: bool = False, verbose: bool = False) -> list:
    """Process and save data"""
    processed_dir = data_dir / "processed"
    processed_dir.mkdir(exist_ok=True)

    processed_paths = []
    skipped_count = 0

    for dicom_folder in tqdm(dicom_folders, desc="Processing data"):
        series_id = dicom_folder.name

        # Save paths
        volume_path = processed_dir / f"{series_id}_volume.npy"
        seg_path = processed_dir / f"{series_id}_segmentation.npy"

        if volume_path.exists() and seg_path.exists() and not force_reprocess:
            # Already processed
            processed_paths.append({
                'series_id': series_id,
                'volume_path': str(volume_path),
                'seg_path': str(seg_path),
                'dicom_folder': str(dicom_folder)
            })
            skipped_count += 1
            continue

        # This is just for tracking
        processed_paths.append({
            'series_id': series_id,
            'volume_path': str(volume_path),
            'seg_path': str(seg_path),
            'dicom_folder': str(dicom_folder)
        })

    print(f"\n✓ Data processing complete:")
    print(f"   - Skipped (already processed): {skipped_count}")
    print(f"   - Total available: {len(processed_paths)} series")

    return processed_paths


def get_augmentation_pipeline():
    """Get data augmentation pipeline"""
    if not AUG_AVAILABLE:
        return None

    return A.Compose([
        A.RandomRotate90(p=0.5),
        A.Flip(p=0.5),
        A.Transpose(p=0.5),
        A.GaussNoise(var_limit=(0.01, 0.05), p=0.2),
        A.RandomBrightnessContrast(brightness_limit=0.1, contrast_limit=0.1, p=0.3),
    ])


def main():
    """Main training function"""
    parser = argparse.ArgumentParser(description='Brain CT Triage Challenge Enhanced Training')
    parser.add_argument('--mode', type=str, default='normal',
                       choices=['normal', 'cv'],
                       help='Training mode: normal or cross-validation')
    parser.add_argument('--epochs', type=int, default=50,
                       help='Number of epochs')
    parser.add_argument('--batch_size', type=int, default=2,
                       help='Batch size')
    parser.add_argument('--lr', type=float, default=1e-4,
                       help='Learning rate')
    parser.add_argument('--num_samples', type=int, default=None,
                       help='Number of samples to use (for testing)')
    parser.add_argument('--force_reprocess', action='store_true',
                       help='Force reprocessing of all data')
    parser.add_argument('--verbose', action='store_true',
                       help='Print verbose output')

    # آرگومان‌های جدید
    parser.add_argument('--model_type', type=str, default='unet',
                       choices=['unet', 'attention_unet', 'deeplabv3'],
                       help='Model architecture to use')
    parser.add_argument('--use_attention', action='store_true',
                       help='Use attention mechanism')
    parser.add_argument('--use_augmentation', action='store_true',
                       help='Use data augmentation')
    parser.add_argument('--scheduler', type=str, default='cosine',
                       choices=['cosine', 'cosine_warm'],
                       help='Learning rate scheduler')
    parser.add_argument('--use_advanced_loss', action='store_true',
                       help='Use combined loss')

    args = parser.parse_args()

    print("\n" + "="*60)
    print("IAAA COMPETITION 2026 - BRAIN CT TRIAGE CHALLENGE")
    print("="*60)
    print(f"Mode: {args.mode}")
    print(f"Epochs: {args.epochs}")
    print(f"Batch size: {args.batch_size}")
    print(f"Learning rate: {args.lr}")
    print(f"Model type: {args.model_type}")
    print(f"Use attention: {args.use_attention}")
    print(f"Use augmentation: {args.use_augmentation}")
    print(f"Scheduler: {args.scheduler}")
    print(f"Use advanced loss: {args.use_advanced_loss}")

    if args.num_samples:
        print(f"Using {args.num_samples} samples (debug mode)")

    # Configuration
    config = {
        'model_type': args.model_type,
        'use_attention': args.use_attention,
        'model_params': {
            'in_channels': 1,
            'seg_classes': 6,
            'init_features': 32
        },
        'learning_rate': args.lr,
        'weight_decay': 1e-5,
        'batch_size': args.batch_size,
        'num_epochs': args.epochs,
        'patience': 15,
        'seg_weight': 1.0,
        'mls_weight': 0.5,
        'fracture_weight': 0.3,
        'dataset_params': {
            'target_size': (256, 256),
            'num_slices': 32,
            'augmentation': args.use_augmentation
        },
        'device': 'cuda' if torch.cuda.is_available() else 'cpu',
        'num_workers': 2,
        'pin_memory': False,
        'use_wandb': False,
        'checkpoint_dir': 'checkpoints',
        'scheduler': args.scheduler,
        'use_advanced_loss': args.use_advanced_loss
    }

    print("\n" + "="*60)
    print("BRAIN CT TRIAGE CHALLENGE - TRAINING")
    print("="*60)

    # Print configuration
    print("\n📋 CONFIGURATION:")
    for key, value in config.items():
        if isinstance(value, dict):
            print(f"   {key}:")
            for k, v in value.items():
                print(f"      {k}: {v}")
        else:
            print(f"   {key}: {value}")

    # 1. Load metadata
    print("\n📊 LOADING METADATA...")
    data_dir = Path("data")
    metadata = load_metadata(data_dir)

    if metadata is None:
        print("✗ No metadata found. Exiting...")
        return

    # 2. Prepare data paths
    print("\n🔍 PREPARING DATA PATHS...")
    dicom_folders = prepare_data_paths(data_dir, metadata, args.num_samples)

    if not dicom_folders:
        print("✗ No valid DICOM folders found. Exiting...")
        return

    # 3. Process data
    print("\n⚙️ PROCESSING DATA...")
    processed_paths = prepare_processed_data(data_dir, dicom_folders, args.force_reprocess, args.verbose)

    if len(processed_paths) == 0:
        print("✗ No data processed successfully. Exiting...")
        return

    if args.mode == 'normal':
        # 4. Train/val split
        print("\n📦 SPLITTING DATA...")
        train_paths, val_paths = train_test_split(
            processed_paths,
            test_size=0.15,
            random_state=42,
            shuffle=True
        )

        print(f"   Train: {len(train_paths)} series")
        print(f"   Val: {len(val_paths)} series")

        # 5. Create datasets
        print("\n🎯 CREATING DATASETS...")

        train_dataset = CTDataset3D(
            dicom_dirs=[p['dicom_folder'] for p in train_paths],
            annotation_dirs=[p['dicom_folder'] for p in train_paths],
            metadata_df=metadata,
            is_training=True,
            processed_dir="data/processed",
            slice_based=True,
            **config['dataset_params']
        )

        val_dataset = CTDataset3D(
            dicom_dirs=[p['dicom_folder'] for p in val_paths],
            annotation_dirs=[p['dicom_folder'] for p in val_paths],
            metadata_df=metadata,
            is_training=False,
            processed_dir="data/processed",
            slice_based=True,
            **config['dataset_params']
        )

        # 6. Create dataloaders
        train_loader = torch.utils.data.DataLoader(
            train_dataset,
            batch_size=config['batch_size'],
            shuffle=True,
            num_workers=config['num_workers'],
            pin_memory=config['pin_memory'],
            drop_last=True
        )

        val_loader = torch.utils.data.DataLoader(
            val_dataset,
            batch_size=config['batch_size'],
            shuffle=False,
            num_workers=config['num_workers'],
            pin_memory=config['pin_memory']
        )

        # 7. Create model
        print("\n🧠 CREATING MODEL...")

        # Always use MultiTaskUNet since other models aren't available
        model = MultiTaskUNet(**config['model_params'])
        print("✅ Using MultiTaskUNet")

        total_params = sum(p.numel() for p in model.parameters())
        trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
        print(f"   Total parameters: {total_params:,}")
        print(f"   Trainable parameters: {trainable_params:,}")

        # 8. Create trainer and train
        print("\n🚀 STARTING TRAINING...")
        trainer = BrainCTTriageTrainer(
            model=model,
            train_loader=train_loader,
            val_loader=val_loader,
            config=config
        )

        train_losses, val_losses = trainer.train(config['num_epochs'])

        # 9. Save final model
        print("\n💾 SAVING FINAL MODEL...")
        weights_dir = Path("weights")
        weights_dir.mkdir(exist_ok=True)

        model_name = "enhanced_model"
        final_model_path = weights_dir / f"{model_name}.pth"
        torch.save({
            'model_state_dict': trainer.best_model_state,
            'config': config,
            'train_losses': train_losses,
            'val_losses': val_losses,
            'best_val_loss': trainer.best_val_loss,
            'epoch': len(train_losses)
        }, final_model_path)

        print(f"✓ Model saved to {final_model_path}")
        print(f"✓ Best validation loss: {trainer.best_val_loss:.4f}")

        # 10. Save training config
        config_path = weights_dir / f"{model_name}_config.json"
        with open(config_path, 'w') as f:
            serializable_config = {}
            for k, v in config.items():
                if isinstance(v, Path):
                    serializable_config[k] = str(v)
                elif isinstance(v, tuple):
                    serializable_config[k] = list(v)
                elif isinstance(v, dict):
                    serializable_config[k] = {kk: str(vv) if isinstance(vv, Path) else vv
                                             for kk, vv in v.items()}
                else:
                    serializable_config[k] = v
            json.dump(serializable_config, f, indent=2)

        print(f"✓ Config saved to {config_path}")

        print("\n" + "="*60)
        print("✅ TRAINING COMPLETED SUCCESSFULLY!")
        print("="*60)

    else:  # Cross-validation mode
        print("\n📊 CROSS-VALIDATION MODE")
        print("   (To be implemented in next version)")


if __name__ == "__main__":
    main()
```
